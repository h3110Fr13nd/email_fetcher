```python
import subprocess
import os
import sys
import pandas as pd
import glob
```

```python
folder_name = "data/german_mails"
output_folder = "data/german_mails/emails"
```

```python
df = pd.read_excel("data/german_mails/Germany Data.xlsx", header=None, names=["domain"])
df
```

```python
# Partition in 30 csv files
n = 150
partition = len(df) // n
for i in range(n):
    df.iloc[i*partition:(i+1)*partition].to_csv(f"{folder_name}/domains/domains_{i}.csv", index=False)
    
```

```python
def run_crawl(start_url_csv, folder_name,i):
    subprocess.Popen(f'''cd {os.getcwd()} && \
                    scrapy crawl fetch_emails_simple \
                    -a start_url_csv="{start_url_csv}" \
                    -o "{folder_name}/emails/emails_{i}.csv" \
                    --logfile "{folder_name}/logs/log_{i}.txt"''',
                    shell=True)
```

```python
csv_files = glob.glob(f"{folder_name}/domains/domains*.csv")
print(csv_files)
for i, csv_file in enumerate(csv_files):
    run_crawl(csv_file, folder_name, i)
```

```python
csv_files = glob.glob(f"{folder_name}/emails/emails*.csv")
len(csv_files)
```

```python
csv_file_list = []
for f in csv_files:
    print(f)
    csv_file_list.append(pd.read_csv(f))
    
df = pd.concat(csv_file_list, ignore_index=True)
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["emails"])
df
```

```python
df.shape
```

```python
df.nunique()
```

```python
df = df.assign(emails=df.emails.str.split(",")).explode("emails")
df
```

```python
df['emails'] = df['emails'].str.strip().str.lower()
df
```

```python
df = df.drop_duplicates(subset=["emails"])
df
```

```python
# get the rows where the emails column contains the value of domain column
df_filtered = df[df.apply(lambda row: row['domain'] in row['emails'], axis=1)]
df_filtered
```

```python
df_rest = df[~df.apply(lambda row: row['domain'] in row['emails'], axis=1)]
df_rest
```

```python
df_filtered.nunique()
```

```python
# reset index to "domain" column
df_filtered = df_filtered.set_index("domain")
```

```python
df_filtered
```

```python
# drop origin_url column
df_filtered.drop(columns=['origin_url'], inplace=True)
```

```python
df_filtered.to_csv(f"{folder_name}/german_clean.csv")
```

```python
df_filtered.drop(columns=['url'], inplace=True)
df_filtered
```

```python
# implode emails with same domain column. Make it comma separated mails in email columns.
df_filtered = df_filtered.groupby('domain').agg({
    'emails' : lambda x: ','.join(list(x.unique()))
}).reset_index()
df_filtered
```
