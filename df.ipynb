```python
import pandas as pd
```

```python
# Get all the csv file names in the current directory starting with "emails"
import glob
csv_files = glob.glob("data/goodfirms_appdev/emails/goodfirms_appdev*.csv")
len(csv_files)
```

```python
# Concatenate all the csv files into one dataframe
csv_file_list = []
for f in csv_files:
    print(f)
    csv_file_list.append(pd.read_csv(f))
    
df = pd.concat(csv_file_list, ignore_index=True)
df
```

```python
df.loc[:,"company_name"] = df["company_name"].str.strip()
df = df.drop_duplicates()
df
```

```python
# drop rows wiith missing email addresses
df = df.dropna(subset=["emails"])
df
```

```python
df.nunique()
```

```python
# split emails into separate rows
df = df.assign(emails=df.emails.str.split(",")).explode("emails")
df
```

```python
# drop duplicate emails
df = df.drop_duplicates(subset=["emails","country"])
df
```

```python
df['country'] = df['country'].str.split(',')
df = df.explode('country')
df['country'] = df['country'].str.strip()
df.nunique()
```

```python
df = df.groupby('emails').agg({
    'country': lambda x: list(x.unique()),
    # company_name to set first non-null value
    'company_name': lambda x: next((i for i in x if pd.notnull(i)), ''),
    'url': lambda x: next((i for i in x if pd.notnull(i)), ''),
}).reset_index()
df
```

```python
df.dtypes
```

```python
# a function to get domain from url
from urllib.parse import urlparse

def get_domain(url):
    if url:
        parsed_uri = urlparse(url)
        domain = parsed_uri.netloc
        return domain
    else:
        return url

df['domain'] = df['url'].apply(get_domain)
df
```

```python
import tldextract

def get_domain_without_subdomain(url):
    extracted = tldextract.extract(url)
    return "{}.{}".format(extracted.domain, extracted.suffix)

df['domain_tld'] = df['url'].apply(get_domain_without_subdomain)
df
```

```python
def get_domain_from_email(email):
    try:
        return email.split('@')[1]
    except IndexError:
        return None
    
df.loc[:,'domain_email'] = df['emails'].apply(get_domain_from_email)
df
```

```python
# keep only rows where domain_email and domain_tld are the same
df = df[df['domain_email'] == df['domain_tld']]
df
```

```python
# drop domain_email and domain_tld columns
df = df.drop(columns=['domain_email', 'domain_tld'])
df
```

```python
df.drop(columns='country').nunique()
# df
```

```python
# reorder columns as company_name, url, emails, country
df = df[['company_name', 'url', 'emails', 'country']]
```

```python
# check if emails are not included in previous emails list
prev_emails_csvs = [
    "data/themanifest_webdev_usuk/emails/emails_clean.csv",
    "data/designrush_webdev_usuk/emails/emails_clean.csv",
]
df_prev = pd.concat([pd.read_csv(f) for f in prev_emails_csvs])

df_prev
```

```python
df_prev['emails'].isin(df['emails']).sum()
```

```python
# remove emails that are already in previous emails list
df = df[~df['emails'].isin(df_prev['emails'])]
df
```
